```python
import sqlite3
import pandas as pd

# Load datasets
athletes = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/athlete_events.csv")
regions = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/noc_regions.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep='\t')


# Connect to SQLite in-memory DB
conn = sqlite3.connect(":memory:")

# Write DataFrames to SQL tables
athletes.to_sql("athletes_table", conn, index=False, if_exists="replace")
regions.to_sql("regions_table", conn, index=False, if_exists="replace")
sales.to_sql("sales_table", conn, index=False, if_exists="replace")
```

```python
# Query to count the total number of medals won by each country.
# Filters out rows where Medal is NULL, groups by NOC (country code),and shows the top 5 countries by total medal count.
pd.read_sql(
    """
    SELECT NOC,
    COUNT(*) AS Total_medals
    FROM athletes_table
    WHERE Medal IS NOT NULL
    GROUP BY NOC
    ORDER BY Total_medals DESC
    LIMIT 5;
    """,
    conn
 )
```

```python
# Query to calculate the average age of athletes who received a Gold medal.
# Filters rows to only include Gold medalists and computes the average of the Age column.
pd.read_sql(
    """
    SELECT avg(Age)
    FROM athletes_table
    WHERE Medal = 'Gold';
    """,
    conn
)
```

```python
# Query to count how many distinct events exist for each sport.
# Uses GROUP BY on Sport and counts DISTINCT Event values.
pd.read_sql(
    """
    SELECT Sport,
    COUNT(DISTINCT Event) AS Total_Distinct_Events
    FROM athletes_table
    GROUP BY Sport
    ORDER BY Total_Distinct_Events DESC;
    """,
    conn
)
```

```python
# Query to retrieve all athlete records where the country code (NOC) is 'USA', indicating that the athlete represents the United States.
pd.read_sql(
    """
    SELECT * FROM athletes_table
    WHERE NOC = 'USA';
    """,
    conn
)
```

```python
# Query to count the number of medals awarded in each Olympic year.
# Filters out rows where Medal is NULL and groups the results by Year.

pd.read_sql(
    """
    SELECT Year,
    COUNT(*) AS Total_medals
    FROM athletes_table
    WHERE Medal IS NOT NULL
    GROUP BY Year
    ORDER BY Total_medals DESC;
    """,
    conn
)

```

```python
# Query to identify athlete records where either height or weight is NULL. This helps in cleaning or imputing missing physical data.
pd.read_sql(
    """
    SELECT * FROM athletes_table
    WHERE Height IS NULL OR Weight IS NULL;
    """,
    conn
)

```

```python
# Query to update missing height values in the athletes table.
# Calculates the average height from non-null records and sets it as the value for any record where height is currently NULL.
cursor = conn.cursor()

cursor.execute("""
    WITH AvgHeight AS (
        SELECT AVG(height) AS avg_height
        FROM athletes_table
        WHERE height IS NOT NULL
    )
    UPDATE athletes_table
    SET height = (SELECT avg_height FROM AvgHeight)
    WHERE height IS NULL;
""")

# Commit the changes
conn.commit()
# Query to verify the update
pd.read_sql("SELECT * FROM athletes_table WHERE height IS  NOT NULL", conn)
```

```python
# Query to calculate the total sales for each item.
# It removes the dollar sign from item_price using REPLACE and casts it as a numeric type.Then multiplies it by quantity and groups by item_name.
pd.read_sql(
    """
    SELECT item_name,
    SUM(CAST(REPLACE(item_price, '$', '') AS DECIMAL(10,2)) * quantity) AS total_sales
    FROM sales_table
    GROUP BY item_name
    ORDER BY total_sales DESC;
    """,
    conn
)
```

```python
# Query to return the top 5 most expensive individual items.
# Casts item_price to a numeric format after removing the dollar sign,orders results in descending order, and limits the output to 5 rows.
pd.read_sql(
    """
    SELECT Item_name,
    CAST(REPLACE(item_price, '$', '') AS REAL) AS item_price
    FROM sales_table
    ORDER BY item_price DESC
    LIMIT 5;
    """,
    conn
)
```

```python
# This query calculates the total number of unique customer orders by counting the distinct values in the order_id column.
pd.read_sql(
    """
    SELECT COUNT(DISTINCT order_id) AS unique_orders
    FROM sales_table
    """,
    conn
)
```

```python
# Query to filter medal-winning athletes and group by country
pd.read_sql(
    """
    WITH medalists_per_country AS (
    SELECT
        NOC,
        COUNT(DISTINCT name) AS medalist_count,
        AVG(age) AS avg_age
    FROM
        athletes_table
    WHERE
        medal IS NOT NULL
        AND age IS NOT NULL
    GROUP BY
        NOC
)
SELECT
    NOC,
    medalist_count,
    ROUND(avg_age, 2) AS average_age,

    -- Classify performance using CASE
    CASE
        WHEN avg_age < 25 THEN 'High'
        WHEN avg_age BETWEEN 25 AND 30 THEN 'Medium'
        ELSE 'Low'
    END AS performance

FROM
    medalists_per_country
ORDER BY
    medalist_count DESC;

    """,
    conn
)
```
